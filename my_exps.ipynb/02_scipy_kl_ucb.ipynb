```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
total = 100
pos = 70
full_count = 1000
coef = 0.5

def d(p, q):
    return p * np.log(p/q) + (1-p) * np.log((1-p)/(1-q))

def f(q):
    rhs = np.log(full_count) + coef * np.log(np.log(full_count))
    return total * d(pos/total, q) - rhs
```

```python
x = np.linspace(0.0, 1.0, 102)[1:-1]
y = f(x)
zero = np.zeros_like(x)
p = pos/total
```

```python
root = optimize.brentq(f, p, 1-1e-12)
print(root)
print(f(root))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, zero)
ax.vlines(p, y.min(), y.max(), linestyles='dotted')
ax.plot(root, f(root), marker="o")
```
